```python
import pandas as pd

df = pd.read_csv("/Users/kethanashambula/Downloads/google_play_store_dataset.csv")

df.head()
```

```python
print("Rows and Columns:", df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Fill Rating with median
df["Rating"] = df["Rating"].fillna(df["Rating"].median())

# Fill categorical columns with mode
df["Type"] = df["Type"].fillna(df["Type"].mode()[0])
df["Content Rating"] = df["Content Rating"].fillna(df["Content Rating"].mode()[0])
df["Current Ver"] = df["Current Ver"].fillna(df["Current Ver"].mode()[0])
df["Android Ver"] = df["Android Ver"].fillna(df["Android Ver"].mode()[0])

# Verify
df.isnull().sum()
```

```python
print("Duplicate rows:", df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True)
```

```python
import matplotlib.pyplot as plt

df["Category"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 App Categories")
plt.xlabel("Category")
plt.ylabel("Number of Apps")

plt.show()
```

```python
top_rated = df.groupby("Category")["Rating"].mean().sort_values(ascending=False).head(10)

top_rated.plot(kind="bar")

plt.title("Top Rated Categories")
plt.xlabel("Category")
plt.ylabel("Average Rating")

plt.show()
```
