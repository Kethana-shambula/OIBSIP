```python
import pandas as pd

df = pd.read_csv("/Users/kethanashambula/Downloads/SampleSuperstore.csv")
df.head()
```

```python
print("Rows and Columns:", df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

category_sales = df.groupby("Category")["Sales"].sum()

plt.figure(figsize=(6,4))
category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.ylabel("Sales")
plt.xlabel("Category")

plt.show()
```

```python
profit_category = df.groupby("Category")["Profit"].sum()

plt.figure(figsize=(6,4))
profit_category.plot(kind="bar")

plt.title("Profit by Category")
plt.ylabel("Profit")
plt.xlabel("Category")

plt.show()
```

```python
region_sales = df.groupby("Region")["Sales"].sum()

plt.figure(figsize=(6,4))
region_sales.plot(kind="bar")

plt.title("Sales by Region")
plt.ylabel("Sales")
plt.xlabel("Region")

plt.show()
```

```python
top_states = df.groupby("State")["Sales"].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_states.plot(kind="bar")

plt.title("Top 10 States by Sales")
plt.ylabel("Sales")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

numeric_df = df[["Sales", "Quantity", "Discount", "Profit"]]

plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
profit_subcategory = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)

plt.figure(figsize=(12,6))
profit_subcategory.head(10).plot(kind="bar")

plt.title("Top 10 Sub-Categories by Profit")
plt.ylabel("Profit")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["Sales"], bins=30)

plt.title("Distribution of Sales")
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```
